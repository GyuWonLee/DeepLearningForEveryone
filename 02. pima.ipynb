```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.callbacks import TensorBoard
from time import time
```

```python
X = df.iloc[:, :8]
y = df.iloc[:, 8]
```

```python
print(type(X), type(y))
```

```python
model = Sequential()
model.add(layers.Dense(12, input_dim=8, activation='relu', name = 'Dense_1'))
model.add(layers.Dense(8 , activation='relu', name = 'Dense_2'))
model.add(layers.Dense(1, activation='sigmoid', name = 'Dense_3'))
```

```python
model.summary()
```

```python
tensorboard = TensorBoard(log_dir = 'logs/{}'.format(time()))
```

```python
model.compile(loss="binary_crossentropy", optimizer='adam', metrics=['accuracy'])
history = model.fit(X, y, epochs=100, batch_size = 5, callbacks = [tensorboard])
```

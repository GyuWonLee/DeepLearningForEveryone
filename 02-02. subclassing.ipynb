```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
```

```python
df = pd.read_csv('data/pima-indians-diabetes3.csv', delimiter=',')
X = df.iloc[:, :8]
y = df.iloc[:, 8]
```

```python
model_name = 'indians_diabetes'
batch_size = 5
params = {'model_name' : model_name,
         'input_dimension' : 8,
         'hidden_dimension1' : 12,
         'hidden_dimension2' : 8, 
         'output_dimension' : 1}
```

```python
class IndiansDiabetes(tf.keras.Model):
    def __init__(self, **kargs):
        super(IndiansDiabetes, self).__init__(name=kargs['model_name'])
        self.fc1 = layers.Dense(kargs['hidden_dimension1'], 
                               input_dim=kargs['input_dimension'],
                               activation='relu')
        self.fc2 = layers.Dense(kargs['hidden_dimension2'], 
                               activation='relu')
        self.fc3 = layers.Dense(kargs['output_dimension'],
                               activation='sigmoid')
        
        
    def call(self, x):
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        return x
```

```python
model = IndiansDiabetes(**params)
```

```python
model.build((batch_size, params['input_dimension']))
```

```python
model.summary()
```

```python
model.compile(loss="binary_crossentropy", optimizer='adam', metrics=['accuracy'])
history = model.fit(X, y, epochs=100, batch_size = 5)
```
